# Persistent Volume (pv) v Persistent Volume Claim (pvc)

PersistentVolume (pv) l mt phn khng gian lu tr d liu trong cluster, cc PersistentVolume ging vi Volume bnh thng tuy nhin n tn ti c lp vi POD (pod b xa PV vn tn ti), c nhiu loi PersistentVolume c th trin khai nh NFS, Clusterfs ...

PersistentVolumeClaim (pvc) l yu cu s dng khng gian lu tr (s dng PV). Hnh dung PV ging nh Node, PVC ging nh POD. POD chy n s dng cc ti nguyn ca NODE, PVC hot ng n s dng ti nguyn ca PV



```python
# trin khai
!kubectl apply -f pv.yaml

# lit k cc PV
!kubectl get pv -o wide

# thng tin chi tit
!kubectl describe pv/pv1
```

## To Persistent Volume Claim trong Kubernetes

PVC (Persistent Volume Claim) l yu cu truy cp n PV, mt PV ch c mt PVC

```python
# trin khai
!kubectl apply -f pvc.yaml

!kubectl get pvc,pv -o wide

!kubectl describe pvc/pvc1
```

## S dng PVC vi Pod

```python
volumes:
# Khai bo VL s dng PVC
- name: myvolume
persistentVolumeClaim:
    claimName: pvc1
containers:
    volumeMounts:
    - mountPath: "/data"
        name: myvolume

```
